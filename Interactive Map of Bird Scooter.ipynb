```python
import folium
from folium.plugins import MarkerCluster
import pandas as pd
import numpy as np

city = ['DC', 38.89511, -77.03637] 
lat = city[1]
lon = city[2]
austin_coords = [lat, lon]
my_map = folium.Map(location = austin_coords, zoom_start = 13)
```

```python
df = pd.read_csv('DC_2.csv', index_col=0)
```

```python
df.head()
```

```python
df['latlon'] = list(zip(df.latitude, df.longitude))
```

```python
coords = [df['latitude'][0], df['longitude'][0]]
```

```python
coords
```

```python
for i in range(0,len(df)):
    folium.Marker([df.iloc[i]['latitude'], df.iloc[i]['longitude']], popup=df.iloc[i]['id']).add_to(my_map)
```
